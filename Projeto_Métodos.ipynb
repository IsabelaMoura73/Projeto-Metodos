## Introdução e Objetivos

O coração humano é um sistema auto oscilante, capaz de gerar impulsos elétricos que regulam de maneira autônoma o ritmo cardíaco. Esses impulsos surgem a partir de variações no potencial de membrana das células cardíacas, ocorrendo de maneira rítmica, embora não perfeitamente regular. Dois tipos celulares especializados com comportamentos distintos sustentam essa atividade, sendo eles:
1.	**Células marca-passo**, que atuam como o principal gerador do ritmo cardíaco. Elas se despolarizam espontaneamente, iniciando cada batimento.
2.	**Células miocárdicas**, que são responsáveis pela propagação do impulso e contração muscular. Diferentemente das células marca-passo, elas não se ativam de forma autônoma, mas respondem aos estímulos gerados por estas.

image.png

A modelagem matemática de células eletricamente excitáveis teve um avanço decisivo com o trabalho pioneiro de **Hodgkin** e **Huxley**, publicado em 1952. Esse modelo, de grande complexidade e vencedor do Prêmio Nobel, representa com alta fidelidade o potencial de ação nos neurônios ao descrever o movimento de íons por canais específicos da membrana celular. Apesar disso, por envolver quatro equações diferenciais acopladas, o modelo possui uma alta complexidade, o que exige análises detalhadas e um elevado custo computacional para simulação.

Dessa forma, para lidar com a complexidade do modelo de Hodgkin-Huxley, tornou-se necessária uma formulação simplificada que preservasse os comportamentos essenciais de uma célula excitável. O modelo de **FitzHugh-Nagumo** atende a essa necessidade, reduzindo o sistema para apenas duas equações. Ele abstrai a complexa dinâmica iônica em uma variável rápida (o potencial de membrana) e uma variável lenta (de recuperação). Essa abordagem captura de forma eficaz aspectos fundamentais como o limiar de excitação, a geração do potencial de ação e o período refratário.

Apesar do modelo ter sido desenvolvido inicialmente para neurônios, a grande aplicabilidade do modelo FHN vem do fato de que células cardíacas e neurônios compartilham a propriedade da excitabilidade elétrica. A ocorrência de um disparo rápido seguido por uma fase de recuperação lenta é um mecanismo universal nesses sistemas. Assim, esse modelo se torna uma ferramenta ideal e eficiente para estudar os mecanismos elétricos do coração, permitindo análises qualitativas de fenômenos complexos como a geração e a propagação de impulsos cardíacos, com baixo custo computacional.

É de grande relevância para as áreas da fisiologia, da bioengenharia e da matemática aplicada representar o potencial de ação cardíaco de forma computacionalmente simples, uma vez que permite a análise qualitativa e quantitativa de fenômenos complexos. O sistema de FitsHugh-Nagumo se destaca por sua simplicidade e eficiência, oferecendo uma maneira acessível de simular e visualizar diferentes padrões oscilatórios relacionados à atividade elétrica do coração.

O projeto possui como objetivo desenvolver, simular e analisar o sistema de FitzHugh-Nagumo sob diferentes funções de forçamento periódico, com o propósito de modelar os potenciais de ação em células cardíacas humanas.
Pretende-se realizar a implementação numérica do sistema por meio do método de Runge-Kutta de quarta ordem para simulação temporal. Além disso, será fundamental testar três diferentes formas de estímulo elétrico $I(t)$ ao sistema:

Essa abordagem combinada de modelagem matemática e análise numérica proporcionará compreensões fundamentais sobre os mecanismos eletrofisiológicos que regulam o ritmo cardíaco.


Em tecidos cardíacos, a comunicação entre células é mediada por estruturas chamadas junções comunicantes (gap junctions), que permitem a passagem direta de íons e pequenas moléculas. Isso possibilita que o potencial de ação se propague rapidamente e de forma coordenada, garantindo a contração eficiente do coração. No modelo de FitzHugh-Nagumo acoplado, o parâmetro

## Modelo

O modelo de FitzHugh–Nagumo é um sistema acoplado de equações diferenciais ordinárias que descreve o comportamento de células excitáveis, como neurônios e células cardíacas. É uma simplificação qualitativa do modelo de Hodgkin–Huxley, reduzido a apenas duas variáveis dinâmicas.

A adoção do modelo simplificado de FitzHugh-Nagumo, em vez do modelo completo de Hodgkin-Huxley, representa uma escolha metodológica intencional. Apesar de o segundo oferecer uma descrição mais detalhada dos mecanismos iônicos, envolvendo um sistema de quatro equações diferenciais para uma única célula, sua elevada complexidade dificultaria a análise matemática. O FitzHugh-Nagumo, composto por apenas duas equações, preserva as propriedades qualitativas essenciais de um sistema excitável, permitindo uma abordagem mais acessível e eficiente do ponto de vista computacional, sendo ideal para os objetivos de análise fundamental deste projeto.

image.png

Para esse escolpo, adotamos a forma forçada do modelo com o objetivo de simular a resposta de células cardíacas a estímulos elétricos periódicos:

$\dfrac{dx}{dt} = y + x - \frac{x³}{3} + I(t)$


$\dfrac{dy}{dt} = -x + a - b*y\$

Onde:



Variáveis
*   $x(t):$ Potencial da membrana	que representa a voltagem da célula, associada à despolarização e repolarização

*   $y(t)$: Variável de recuperação	que modela os efeitos de processos lentos como reabertura de canais iônicos

*  $t$: Evolução temporal do sistema

Parâmetros

*   $a$: Constante de deslocamento, afeta o ponto de equilíbrio do sistema

*   $b$: Força de feedback sobre y, controla a taxa de recuperação

*  $I(t)$: Estímulo externo aplicado à célula

Apesar do sistema ter três possíveis parâmetros — $a$, $b$ e $I(t)$ — o nosso principal foco é em como $I(t)$ afeta o comportamento do sistema, pois o estímulo externo é fundamental para diferenciar o comportamento entre os dois tipos celulares. Mas, mesmo assim, é fundamental destacar que o estímulo externo $I(t)$ atua de forma independente do acoplamento. Cada célula recebe seu próprio estímulo:


* **Células miocárdicas** não geram impulsos por conta própria; elas respondem ao estímulo vindo do marca-passo. Para modelá-las, usamos um estímulo com onda quadrada, que simula pulsos elétricos intermitentes aplicados à célula.

* **Células do marca-passo** têm atividade oscilatória intrínseca. Para modelá-las, usamos um estímulo senoidal (cossenoidal), que representa o ritmo interno de disparo da célula.



## Resultados e Discussão

Nesta seção, implementamos uma simulação computacional do modelo de FitzHugh–Nagumo com duas células cardíacas acopladas: uma célula marca-passo (com estímulo senoidal) e uma célula miocárdica (com estímulo em forma de onda quadrada). O objetivo é analisar como diferentes níveis de acoplamento elétrico afetam a sincronização e a propagação do potencial de ação entre elas.

Utilizamos o método de Runge-Kutta de quarta ordem para resolver numericamente o sistema de equações diferenciais que descreve a dinâmica das variáveis $x1(t)$ e $x2(t)$, correspondentes ao potencial de membrana das duas células. O acoplamento entre elas é modelado por um parâmetro $K$, que representa a intensidade da interação elétrica.

As equações que representam o acoplamento das duas células cardíacas são dadas por:

\begin{cases} \frac{dx_1}{dt} = y_1 + x_1 - \frac{x_1^3}{3} + I_1(t) + K(x_2 - x_1) \\ \frac{dy_1}{dt} = -x_1 + a_1 - b_1 y_1 \\ \frac{dx_2}{dt} = y_2 + x_2 - \frac{x_2^3}{3} + I_2(t) + K(x_1 - x_2) \\ \frac{dy_2}{dt} = -x_2 + a_2 - b_2 y_2 \end{cases}

Onde:

**x1 e x2:** representam o potencial de ação da célula 1 ou 2.

**y1 e y2:** representam a variável de recuperação da célula 1 ou 2.

**K(xj - xi):** representam a força que "puxa" a célula $i$ para se alinhar com a célula $j$. Isso representa a comunicação elétrica entre as células.

Assim, mesmo que K=0, as células ainda apresentam comportamento independente baseado em seus estímulos individuais. O parâmetro $K$ apenas regula a intensidade da interação entre elas, permitindo analisar o impacto da comunicação elétrica sobre a sincronização do sistema.

Para avaliar a importância crucial da comunicação intercelular, analisamos o comportamento do sistema para três cenários distintos de acoplamento:

**K = 0.08 (Acoplamento Fraco):** Simula uma condição de comunicação deficiente ou patológica.

**K = 2.0 (Acoplamento Intermediário):** Representa um limiar onde a sincronização começa a emergir.

**K = 5.0 (Acoplamento Forte):** Modela um tecido cardíaco saudável com comunicação eficiente.

Os gráficos a seguir mostram:

*  A evolução temporal dos potenciais de ação das duas células.

*  A diferença entre os potenciais ao longo do tempo, evidenciando o grau de sincronização.

* o retrato de fase x1 X x2, que ilustra visualmente a convergência das dinâmicas.

Esses resultados permitem avaliar como o sistema responde para diferentes valores de K, modelando desde casos de isolamento elétrico até sincronização completa, refletindo situações fisiológicas e patológicas do tecido cardíaco.



```python
# Bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sistema de equações acopladas
def fhn_coupled(t, X, a1, b1, a2, b2, A1, A2, w1, w2, K):
    x1, y1, x2, y2 = X

    # Estímulo da célula 1 (marca-passo): cosseno
    I1 = -A1 * np.cos(w1 * t)

    # Estímulo da célula 2 (miocárdica): onda quadrada
    I2 = -A2 * np.sign(np.sin(w2 * t))

    # Equações diferenciais com acoplamento K
    dx1 = y1 + x1 - (x1**3)/3 + I1 + K*(x2 - x1)
    dy1 = -x1 + a1 - b1*y1

    dx2 = y2 + x2 - (x2**3)/3 + I2 + K*(x1 - x2)
    dy2 = -x2 + a2 - b2*y2

    return np.array([dx1, dy1, dx2, dy2])
```

```python
# Método de Runge-Kutta de 4º ordem para integrar numericamente o sistema de equações
def rk4(f, X0, t0, tf, dt, *args):
    N = int((tf - t0) / dt)         # número de passos
    t = np.linspace(t0, tf, N + 1)  # vetor de tempo
    X = np.zeros((N + 1, len(X0)))  # matriz para armazenar as variáveis

    X[0] = X0  # condições iniciais

    for i in range(N):
        k1 = f(t[i], X[i], *args)
        k2 = f(t[i] + dt / 2, X[i] + dt * k1 / 2, *args)
        k3 = f(t[i] + dt / 2, X[i] + dt * k2 / 2, *args)
        k4 = f(t[i] + dt, X[i] + dt * k3, *args)
        X[i + 1] = X[i] + (dt / 6) * (k1 + 2 * k2 + 2 * k3 + k4)

    return t, X
```

### **K = 0.08**

```python
# Definindo os parâmetros (a, b, A e w vem do artigo)
# Parâmetros da célula marca-passo
a1, b1 = 0.9, 0.2

# Parâmetros da célula miocárdica
a2, b2 = 0.6, 0.75

# Parâmetros dos estímulos externos
A1, w1 = 5.0, 0.5   # cos(t) para x1
A2, w2 = 5.0, 0.5   # onda quadrada para x2

# Acoplamento
K = 0.08
```

```python
# Definindo condições iniciais e o intervalo
# Condições iniciais: x1, y1, x2, y2
X0 = [0.0, 0.0, 0.0, 0.0]

# Intervalo de tempo
t0 = 0
tf = 100
dt = 0.05
```

```python
# Executando a simulação
t, X = rk4(fhn_coupled, X0, t0, tf, dt, a1, b1, a2, b2, A1, A2, w1, w2, K)
```

```python
# Separando as variáveis
x1, y1, x2, y2 = X.T  # Transpõe e separa
```

```python
# Plotando os resultados
plt.figure(figsize=(12, 6))
plt.plot(t, x1, label='x1 (marca-passo)', color='deeppink')
plt.plot(t, x2, label='x2 (miocárdio)', linestyle='dashed', color='hotpink')
plt.xlabel('t (ms)')
plt.ylabel('Potencial de ação')
plt.title(f'Sistema Acoplado FitzHugh-Nagumo (K={K})')
plt.legend()
plt.grid()
plt.show()
```

Para um acoplamento fraco de K = 0.08, o modelo demonstra uma falha na sincronização entre a célula marca-passo e a miocárdica. Isso simula uma condição fisiopatológica onde as junções comunicantes são ineficazes ou estão danificadas, impedindo que o impulso elétrico gerado pela célula marca-passo seja transmitido e comande a contração da célula muscular. O resultado é um sistema dessincronizado, onde cada célula segue seu próprio ritmo, o que no coração levaria a uma arritmia severa ou à perda da função de bombeamento.

```python
# Vizualizando a relação de x1 e x2
plt.figure(figsize=(6, 5))
plt.plot(x1, x2, color='deeppink', label='x1 vs x2')
plt.plot(x1, x1, color='gray', linestyle='--', label='linha identidade')  # x1 = x2
plt.xlabel('x1 (marca-passo)')
plt.ylabel('x2 (miocárdio)')
plt.title(f'Relação x1 vs x2 (K={K})')
plt.legend()
plt.grid()
plt.axis('equal')
plt.show()
```

Nesse gráfico, a linha cinza tracejada representa a "linha de sincronia perfeita". Se as duas células estivessem perfeitamente sincronizadas, a trajetória rosa estaria exatamente em cima dessa linha cinza, pois a todo momento teríamos x1(t) = x2(t).
Logo, para um acoplamento fraco de K = 0.08, o sistema é dessincronizado. A relação entre os potenciais das duas células é complexa e não converge para um estado coordenado, o que é a comprovação de um sistema fracamente acoplado.

```python
# Diferença entre as células
plt.figure(figsize=(10, 4))
plt.plot(t, np.abs(x1 - x2), color='deeppink')
plt.xlabel('t (ms)')
plt.ylabel('|x1 - x2|')
plt.title(f'Diferença de potencial entre as células (K={K})')
plt.grid()
plt.show()
```

A linha representa o valor absoluto da diferença entre o potencial da célula 1 e da célula 2 em cada instante. Se as células estiverem perfeitamente sincronizadas, a diferença |x1 - x2| será zero. Se elas estiverem dessincronizadas, a diferença será maior que zero.
Em nenhum momento a diferença entre as células se aproxima de zero e permanece lá. Isso significa que as células estão persistentemente fora de sincronia.

A diferença entre as células não desaparece, mas se estabiliza em uma oscilação periódica, refletindo o atraso de fase constante entre a célula marca-passo e a célula miocárdica. Este é o comportamento característico de um sistema que está "travado" em fase, mas não perfeitamente em fase.

## Conclusão

O presente estudo demonstrou que o modelo acoplado de FitzHugh–Nagumo, mesmo sendo uma formulação simplificada, é capaz de reproduzir qualitativamente comportamentos essenciais do tecido cardíaco, como a geração, propagação e sincronização de potenciais de ação entre células de marca-passo e miocárdicas. Ao implementar o método numérico de Runge–Kutta de quarta ordem, foi possível explorar de forma controlada o papel do parâmetro de acoplamento elétrico K, que modela a eficiência de comunicação via junções comunicantes.

A análise sistemática de três regimes distintos de acoplamento revelou um comportamento de transição bem definido:

**K baixo (0,08)** → Falha de sincronização, representando condições patológicas severas, como bloqueios de condução ou danos estruturais que comprometem as gap junctions.

**K intermediário (2,0)** → Sincronização parcial, reproduzindo cenários de condução elétrica degradada, compatíveis com fibrose inicial, envelhecimento do tecido ou início de arritmias.

**K alto (5,0)** → Sincronização quase perfeita, característica de um tecido saudável, onde o ritmo ditado pela célula marca-passo é fielmente seguido pelas células miocárdicas.

Esses resultados validam a relevância fisiológica e didática do modelo de FitzHugh–Nagumo no estudo da eletrofisiologia cardíaca, especialmente por oferecer baixo custo computacional e facilidade de manipulação de parâmetros, preservando as propriedades qualitativas fundamentais de um sistema excitável. Além disso, a abordagem combinando modelagem matemática, integração numérica e visualização gráfica proporcionou uma compreensão integrada de como a comunicação intercelular afeta a dinâmica coletiva do tecido cardíaco.

A partir dessa base, o trabalho abre espaço para diversos desdobramentos, como a implementação do modelo de Hodgkin–Huxley para uma comparação quantitativa mais rigorosa; a introdução de heterogeneidade celular para simular diferentes tipos ou estados patológicos e a modelagem de acoplamento dinâmico com K variável no tempo.

Assim, este trabalho não apenas alcançou seu objetivo inicial de caracterizar a influência do acoplamento elétrico na sincronização de células cardíacas, mas também estabeleceu um marco conceitual e metodológico que pode ser expandido tanto para aprofundar a compreensão de fenômenos fisiológicos e patológicos quanto para orientar aplicações práticas em cardiologia, fisiologia computacional e engenharia biomédica.
